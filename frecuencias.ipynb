```python
import pandas as pd

frecuencias=pd.read_excel('1inicio.xlsx',skiprows=[0,1])
frecuencias
```

```python
pais=pd.DataFrame(data=frecuencias,columns=('main_country','title'))
pais
```

```python

pais_frecuencia=pais['main_country'].value_counts()
pais_frecuencia
```

```python
pais_frecuencia_relativa=pais['main_country'].value_counts(normalize=True)*100
pais_frecuencia_relativa
```

```python
periodo_estreno=pd.DataFrame(data=frecuencias,columns=('release_year','title'))
periodo_estreno
```

```python
periodo_estreno['decade']=(periodo_estreno['release_year'] // 10)*10
periodo_estreno_decada_frecuencia=periodo_estreno.groupby('decade').size()
periodo_estreno_decada_frecuencia
```
