```python
import pandas as pd
import matplotlib.pyplot as plt

datos=pd.read_excel('1inicio.xlsx',skiprows=[0,1])
```

```python
tabla1=pd.DataFrame(data=datos,columns=('rating','title'))
datos_grafico_barras=tabla1['rating'].value_counts()
datos_grafico_barras
```

```python
datos_grafico_barras.plot(kind='bar',color='skyblue')
plt.xlabel('rating')
plt.ylabel('Cuenta')
plt.title('Rating')
plt.show()
```

```python
tabla2=pd.DataFrame(data=datos,columns=('type','title'))
datos_grafico_pastel=tabla2['type'].value_counts()
datos_grafico_pastel
```

```python
datos_grafico_pastel.plot(kind='pie',title='type')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
tabla3=pd.DataFrame(data=datos,columns=('release_year','title'))
datos_grafico_linea=tabla3['release_year'].value_counts().sort_index()
datos_grafico_linea
```

```python
datos_grafico_linea.plot(kind='line')
plt.xlabel('release year')
plt.ylabel('Cuenta')
plt.show()
```
